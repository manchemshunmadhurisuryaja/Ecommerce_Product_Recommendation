```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
df_content = df[['product_id', 'product_name', 'department', 'reordered']]
df_content['product_text'] = df_content['product_name'] + ' ' + df_content['department']
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

df_content['product_text'] = df_content['product_text'].apply(clean_text)
```

```python
# =========================================================
##TF-IDF model (Content Based Filtering)
# =========================================================
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=2)
tfidf_matrix = tfidf_vectorizer.fit_transform(product_df['product_text'])
cosine_sim_tfidf = cosine_similarity(tfidf_matrix, tfidf_matrix)

product_indices = pd.Series(index=product_df['product_id'], data=range(len(product_df)))

def recommend_products_tfidf(product_id, top_k=5):
    idx = product_indices[product_id]
    sim_scores = list(enumerate(cosine_sim_tfidf[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:top_k+1]
    product_idxs = [i[0] for i in sim_scores]
    return product_df.iloc[product_idxs]['product_text'].values

example_product_id = product_df['product_id'].iloc[2]
example_idx = product_indices[example_product_id]

print("Example Product Text:")
print(product_df.iloc[example_idx]['product_text'])

print("\nTop 5 TF-IDF Recommendations:")
example_recommendations = recommend_products_tfidf(example_product_id, 5)
for i, rec in enumerate(example_recommendations, 1):
    print(i, rec)

user_product = data[['user_id', 'product_id', 'reordered']]

def evaluate_tfidf(user_product, k=5):
    precision_list, recall_list, mrr_list = [], [], []
    for user in user_product['user_id'].unique()[:500]:
        user_data = user_product[user_product['user_id'] == user]
        purchased = user_data[user_data['reordered'] == 1]['product_id'].values
        if len(purchased) < 2:
            continue
        query_product = purchased[0]
        actual_products = purchased[1:]
        if query_product not in product_indices:
            continue
        recommended = recommend_products_tfidf(query_product, k)
        actual_product_texts = [product_df.iloc[product_indices[pid]]['product_text'] for pid in actual_products]
        hits = [1 if item in actual_product_texts else 0 for item in recommended]
        precision_list.append(sum(hits) / k)
        recall_list.append(sum(hits) / len(actual_products))
        if 1 in hits:
            mrr_list.append(1 / (hits.index(1) + 1))
        else:
            mrr_list.append(0)
    return np.mean(precision_list), np.mean(recall_list), np.mean(mrr_list)

precision_tfidf, recall_tfidf, mrr_tfidf = evaluate_tfidf(user_product, 5)

print("\nTF-IDF Evaluation Metrics")
print("Precision@5:", precision_tfidf)
print("Recall@5:", recall_tfidf)
print("MRR:", mrr_tfidf)
```

```python
# =========================================================
##BPR model
# =========================================================
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.model_selection import train_test_split
from implicit.bpr import BayesianPersonalizedRanking

# Encode IDs
user_map = {u:i for i,u in enumerate(interactions_df['user_id'].unique())}
item_map = {i:j for j,i in enumerate(interactions_df['product_id'].unique())}

interactions_df['u'] = interactions_df['user_id'].map(user_map)
interactions_df['i'] = interactions_df['product_id'].map(item_map)

train, test = train_test_split(interactions_df, test_size=0.2, random_state=42)

train_matrix = csr_matrix(
    (train['interaction'], (train['u'], train['i']))
)

model = BayesianPersonalizedRanking(
    factors=50,
    learning_rate=0.01,
    iterations=30
)
model.fit(train_matrix)

def evaluate_bpr(model, train, test, k=5):
    precision, recall, mrr = [], [], []

    for user in test['u'].unique():
        true_items = test[test['u']==user]['i'].values
        recs = model.recommend(user, train_matrix[user], N=k)[0]

        hits = set(recs) & set(true_items)
        precision.append(len(hits)/k)
        recall.append(len(hits)/len(true_items))

        rr = 0
        for rank, item in enumerate(recs, 1):
            if item in true_items:
                rr = 1/rank
                break
        mrr.append(rr)

    return np.mean(precision), np.mean(recall), np.mean(mrr)

p, r, m = evaluate_bpr(model, train, test)
print("BPR -> Precision@5:", p, "Recall@5:", r, "MRR:", m)
```

```python
import pandas as pd
import numpy as np

# Quantity-based popularity
quantity_popularity = df1.groupby('product_id')['add_to_cart_order'].sum().sort_values(ascending=False)

# Reorder-based popularity
reorder_popularity = df1.groupby('product_id')['reordered'].sum().sort_values(ascending=False)

# Time-weighted / trending popularity
# More recent orders are weighted higher
df1['time_weight'] = 1 / (df1['days_since_prior_order'].fillna(0) + 1)
time_popularity = df1.groupby('product_id')['time_weight'].sum().sort_values(ascending=False)

# Hybrid popularity (combine quantity + reorder + time)
# Weighted combination: you can adjust weights as needed
hybrid_popularity = df1.groupby('product_id').apply(
    lambda x: x['add_to_cart_order'].sum() * 0.3 +   # quantity weight
              x['reordered'].sum() * 0.5 +           # reorder weight
              x['time_weight'].sum() * 0.2            # time-weighted/trending weight
).sort_values(ascending=False)


# Store all models in a dictionary
pop_models = {
    "quantity": quantity_popularity,
    "reorder": reorder_popularity,
    "time_weighted": time_popularity,
    "hybrid": hybrid_popularity
}

# 2️⃣ Recommendation function for popularity models
def recommend_popularity(user_id, model_name="quantity", top_k=5):
    popular_products = pop_models[model_name].head(top_k).index.tolist()
    return popular_products

# 3️⃣ Evaluation (Precision@K, Recall@K, MRR)
user_product = df1[['user_id', 'product_id', 'reordered']]

def evaluate_popularity(model_name="quantity", k=5):
    precision_list, recall_list, mrr_list = [], [], []

    for user in user_product['user_id'].unique()[:500]:  # evaluate on first 500 users
        user_data = user_product[user_product['user_id'] == user]
        purchased = user_data[user_data['reordered'] == 1]['product_id'].values

        if len(purchased) == 0:
            continue

        recommended = recommend_popularity(user, model_name, top_k=k)
        hits = [1 if item in purchased else 0 for item in recommended]

        precision_list.append(sum(hits) / k)
        recall_list.append(sum(hits) / len(purchased))

        # MRR
        if 1 in hits:
            mrr_list.append(1 / (hits.index(1) + 1))
        else:
            mrr_list.append(0)

    return (
        np.mean(precision_list),
        np.mean(recall_list),
        np.mean(mrr_list)
    )

# 4️⃣ Run evaluation for all popularity models
for model in pop_models.keys():
    precision, recall, mrr = evaluate_popularity(model, k=5)
    print(f"{model.capitalize()} Popularity Model:")
    print(f"Precision@5: {precision:.4f}, Recall@5: {recall:.4f}, MRR: {mrr:.4f}\n")
```

```python
# =========================================================
# Hybrid combination of Content (TF-IDF), Collaborative & Popularity
# =========================================================

import pandas as pd
import numpy as np
from implicit.bpr import BayesianPersonalizedRanking
from scipy.sparse import coo_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# =========================
# 0️⃣ Interaction & mappings
# =========================
df1['interaction'] = 1 + df1['reordered'] * 2 + 1 / df1['add_to_cart_order']

unique_users = df1['user_id'].unique()
unique_products = df1['product_id'].unique()

user_mapper = {u: i for i, u in enumerate(unique_users)}
product_mapper = {p: i for i, p in enumerate(unique_products)}
reverse_item_mapper = {i: p for p, i in product_mapper.items()}

rows = df1['user_id'].map(user_mapper)
cols = df1['product_id'].map(product_mapper)
data = df1['interaction']

train_matrix = coo_matrix(
    (data, (rows, cols)),
    shape=(len(unique_users), len(unique_products))
).tocsr()

# =========================
# 1️⃣ Train BPR model
# =========================
bpr_model = BayesianPersonalizedRanking(
    factors=50,
    learning_rate=0.01,
    iterations=30
)
bpr_model.fit(train_matrix)

# =========================
# 2️⃣ Popularity model
# =========================
reorder_popularity = (
    df1.groupby('product_id')['reordered']
    .sum()
    .sort_values(ascending=False)
)

# =========================
# 3️⃣ TF-IDF Content Model
# =========================

tfidf = TfidfVectorizer(max_features=5000)
tfidf_matrix = tfidf.fit_transform(product_df['product_text'])

cosine_sim_tfidf = cosine_similarity(tfidf_matrix, tfidf_matrix)

product_indices = pd.Series(
    product_df.index,
    index=product_df['product_id']
)


# =========================
# 4️⃣ Recommendation Functions
# =========================

# ---------- Content-based (TF-IDF) ----------
def recommend_content(user_history_product_ids, top_k=5):
    scores = np.zeros(len(product_df))

    for pid in user_history_product_ids:
        if pid in product_indices:
            scores += cosine_sim_tfidf[product_indices[pid]]

    top_indices = np.argsort(scores)[::-1][:top_k]
    return product_df.iloc[top_indices]['product_id'].tolist()


# ---------- BPR ----------
def recommend_bpr(user_id, top_k=5):
    if user_id not in user_mapper:
        return reorder_popularity.head(top_k).index.tolist()

    user_idx = user_mapper[user_id]
    recs = bpr_model.recommend(
        user_idx,
        train_matrix[user_idx],
        N=top_k,
        filter_already_liked_items=True
    )
    return [reverse_item_mapper[int(i[0])] for i in recs]


# ---------- Popularity ----------
def recommend_popularity(top_k=5):
    return reorder_popularity.head(top_k).index.tolist()


# ---------- Weighted Hybrid ----------
def recommend_hybrid(user_id, top_k=5, w_tfidf=0.4, w_bpr=0.4, w_pop=0.2):
    user_history = df1[df1['user_id'] == user_id]['product_id'].unique()

    content_score = np.zeros(len(product_df))
    for pid in user_history:
        if pid in product_indices:
            content_score += cosine_sim_tfidf[product_indices[pid]]

    bpr_score = np.zeros(len(product_df))
    if user_id in user_mapper:
        recs = bpr_model.recommend(
            user_mapper[user_id],
            train_matrix[user_mapper[user_id]],
            N=len(product_df),
            filter_already_liked_items=False
        )
        for r in recs:
            bpr_score[int(r[0])] = r[1]

    pop_score = reorder_popularity / reorder_popularity.max()

    final_score = (
        w_tfidf * content_score +
        w_bpr * bpr_score +
        w_pop * pop_score.values
    )

    top_indices = np.argsort(final_score)[::-1][:top_k]
    return product_df.iloc[top_indices]['product_id'].tolist()


# ---------- Two-Stage Hybrid ----------
def recommend_two_stage(user_id, top_k=5, candidate_size=100,
                        w_tfidf=0.5, w_bpr=0.5):

    candidates = reorder_popularity.head(candidate_size).index.tolist()
    final_scores = {}

    user_history = df1[df1['user_id'] == user_id]['product_id'].unique()

    for pid in user_history:
        if pid in product_indices:
            sims = cosine_sim_tfidf[product_indices[pid]]
            for idx, score in enumerate(sims):
                prod_id = product_df.iloc[idx]['product_id']
                if prod_id in candidates:
                    final_scores[prod_id] = final_scores.get(prod_id, 0) + w_tfidf * score

    if user_id in user_mapper:
        recs = bpr_model.recommend(
            user_mapper[user_id],
            train_matrix[user_mapper[user_id]],
            N=len(candidates),
            filter_already_liked_items=False
        )
        for r in recs:
            pid = reverse_item_mapper[int(r[0])]
            if pid in candidates:
                final_scores[pid] = final_scores.get(pid, 0) + w_bpr * r[1]

    ranked = sorted(final_scores.items(), key=lambda x: x[1], reverse=True)
    return [pid for pid, _ in ranked[:top_k]]


# ---------- Switching Hybrid ----------
def recommend_switching_hybrid(user_id, top_k=5, min_interactions=5):

    user_history = df1[df1['user_id'] == user_id]['product_id'].unique()

    if len(user_history) < min_interactions or user_id not in user_mapper:
        final = list(dict.fromkeys(
            recommend_content(user_history, top_k * 2) +
            recommend_popularity(top_k * 2)
        ))
        return final[:top_k]

    final = list(dict.fromkeys(
        recommend_bpr(user_id, top_k * 2) +
        recommend_content(user_history, top_k * 2)
    ))
    return final[:top_k]


# =========================
# 5️⃣ Evaluation
# =========================
def evaluate_generic(recommend_fn, user_product, k=5, user_limit=500):
    precision, recall, mrr = [], [], []

    for user in user_product['user_id'].unique()[:user_limit]:
        purchased = user_product[
            (user_product['user_id'] == user) &
            (user_product['reordered'] == 1)
        ]['product_id'].values

        if len(purchased) == 0:
            continue

        recs = recommend_fn(user, k)
        hits = [1 if p in purchased else 0 for p in recs]

        precision.append(sum(hits) / k)
        recall.append(sum(hits) / len(purchased))
        mrr.append(1 / (hits.index(1) + 1) if 1 in hits else 0)

    return np.mean(precision), np.mean(recall), np.mean(mrr)


# =========================
# 6️⃣ Run Experiments
# =========================
user_product = df1[['user_id', 'product_id', 'reordered']]
results = {}

results['TF-IDF'] = evaluate_generic(
    lambda u, k: recommend_content(
        df1[df1['user_id'] == u]['product_id'].unique(), k),
    user_product
)

results['BPR'] = evaluate_generic(recommend_bpr, user_product)
results['Popularity'] = evaluate_generic(lambda u, k: recommend_popularity(k), user_product)
results['Hybrid-1'] = evaluate_generic(lambda u, k: recommend_hybrid(u, k, 0.33, 0.33, 0.34), user_product)
results['Hybrid-2'] = evaluate_generic(lambda u, k: recommend_hybrid(u, k, 0.5, 0.3, 0.2), user_product)
results['Hybrid-3'] = evaluate_generic(lambda u, k: recommend_hybrid(u, k, 0.3, 0.5, 0.2), user_product)
results['Two-Stage Hybrid'] = evaluate_generic(recommend_two_stage, user_product)
results['Switching Hybrid'] = evaluate_generic(recommend_switching_hybrid, user_product)

results_df = pd.DataFrame(results, index=['Precision@5', 'Recall@5', 'MRR']).T
print(results_df)
```

```python
def recommend_two_stage_single(user_id, top_k=5):

    # -------- User Existence Check --------
    if user_id not in df1['user_id'].unique():
        return {
            "status": "not_existing",
            "message": "User does not exist. Please click on New User."
        }

    # -------- Helper: Product Details --------
    def get_product_details(pid):
        row = df1[df1['product_id'] == pid][['product_name', 'department']]
        if not row.empty:
            return {
                "product_id": pid,
                "product_name": row.iloc[0]['product_name'],
                "department": row.iloc[0]['department']
            }
        return None

    user_history = df1[df1['user_id'] == user_id]['product_id'].unique()

    # -------- Cold or Active User --------
    if user_id not in user_mapper or len(user_history) < 5:
        candidate_size = 30
        w_tfidf, w_bpr, w_pop = 0.5, 0.0, 0.5
    else:
        candidate_size = 100
        w_tfidf, w_bpr, w_pop = 0.4, 0.4, 0.2

    candidates = reorder_popularity.head(candidate_size).index.tolist()
    final_scores = {}
    explanations = {}

    # Initialize explanation tracking
    for pid in candidates:
        explanations[pid] = {
            "content_score": 0,
            "collaborative_score": 0,
            "popularity_score": 0
        }

    # -------- TF-IDF Contribution --------
    for hist_pid in user_history:
        if hist_pid in product_indices:
            sims = cosine_sim_tfidf[product_indices[hist_pid]]
            for idx, score in enumerate(sims):
                prod_id = product_df.iloc[idx]['product_id']
                if prod_id in candidates:
                    contrib = w_tfidf * score
                    final_scores[prod_id] = final_scores.get(prod_id, 0) + contrib
                    explanations[prod_id]["content_score"] += contrib

    # -------- BPR Contribution --------
    if user_id in user_mapper and w_bpr > 0:
        user_idx = user_mapper[user_id]
        recs = bpr_model.recommend(
            user_idx,
            train_matrix[user_idx],
            N=len(candidates),
            filter_already_liked_items=False
        )

        for rec in recs:
            item_idx = int(rec[0])
            score = rec[1]
            pid = reverse_item_mapper[item_idx]
            if pid in candidates:
                contrib = w_bpr * score
                final_scores[pid] = final_scores.get(pid, 0) + contrib
                explanations[pid]["collaborative_score"] += contrib

    # -------- Popularity Contribution --------
    pop_score = reorder_popularity / reorder_popularity.max()
    for pid in candidates:
        contrib = w_pop * pop_score[pid]
        final_scores[pid] = final_scores.get(pid, 0) + contrib
        explanations[pid]["popularity_score"] += contrib

    # -------- Ranking --------
    ranked = sorted(final_scores.items(), key=lambda x: x[1], reverse=True)
    top_product_ids = [pid for pid, _ in ranked[:top_k]]

    recommended_items = []

    for pid in top_product_ids:
        details = get_product_details(pid)
        exp = explanations[pid]
        total = sum(exp.values())

        # Percentage calculation
        if total == 0:
            content_pct = collab_pct = pop_pct = 0
        else:
            content_pct = round((exp["content_score"] / total) * 100, 2)
            collab_pct = round((exp["collaborative_score"] / total) * 100, 2)
            pop_pct = round((exp["popularity_score"] / total) * 100, 2)

        # Determine dominant factor
        dominant = max(exp, key=exp.get)

        # -------- Generate Explanation Text --------
        if dominant == "content_score" and len(user_history) > 0:
            similar_items = []

         # Find similarity between recommended product and each past purchased item
            if pid in product_indices:
                rec_idx = product_indices[pid]
    
                for hist_pid in user_history:
                    if hist_pid in product_indices:
                        hist_idx = product_indices[hist_pid]
                        similarity_score = cosine_sim_tfidf[rec_idx][hist_idx]
                        similar_items.append((hist_pid, similarity_score))
    
                # Sort by similarity score
                similar_items = sorted(similar_items, key=lambda x: x[1], reverse=True)
            
                # Take top 2 similar purchased items
                top_similar = similar_items[:2]
            
                hist_names = []
                for item_id, _ in top_similar:
                    name = df1[df1['product_id'] == item_id]['product_name'].iloc[0]
                    hist_names.append(name)
            
                if len(hist_names) == 1:
                    explanation_text = (
                        f"Recommended because it is similar to your previously purchased item: {hist_names[0]}."
                    )
                elif len(hist_names) >= 2:
                    explanation_text = (
                        f"Recommended because it is similar to your previously purchased items: "
                        f"{hist_names[0]} and {hist_names[1]}."
                    )
                else:
                    explanation_text = "Recommended based on your purchase preferences."
            else:
                explanation_text = "Recommended based on your purchase preferences."

        elif dominant == "collaborative_score":
            explanation_text = (
                "Recommended because users with similar purchase patterns also bought this item."
            )

        else:
            explanation_text = (
                "Recommended because this item is currently trending and popular."
            )

        details.update({
            "explanation": explanation_text,
            "contribution_percentages": {
                "content": content_pct,
                "collaborative": collab_pct,
                "popularity": pop_pct
            }
        })

        recommended_items.append(details)

    # -------- Trending Section --------
    trending_ids = reorder_popularity.head(5).index.tolist()
    trending_items = [get_product_details(pid) for pid in trending_ids]

    return {
        "status": "existing_user",
        "recommended": recommended_items,
        "trending": trending_items
    }
```
